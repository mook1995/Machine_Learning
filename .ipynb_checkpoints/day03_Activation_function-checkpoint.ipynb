## 활성화 함수 (activation function)

## 01.Sigmoid 
- y = 1 / (1 + e <sup> -x </sup> )

```python
import numpy as np 
import matplotlib.pyplot as plt

def sigmoid_function(x) :
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-5, 5)
y = sigmoid_function(x)
plt.plot(x,y)
plt.title("Sigmoid")
plt.show()
##시그모이드 함수는 x갑이 무한이 작아질수록 0에 가까워지고 커질수록 1에 가까워지는 함수이다. 
```

- y<sup>'</sup> = (1-y) * y

## 02.  tanh (하이퍼볼릭 탄젠트)
 y = (e<sup>x</sup> - e<sup>-x</sup>) / (e<sup>x</sup> + e<sup>-x</sup>)

```python
def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-5, 5)
y = tanh_function(x)
plt.plot(x,y)
plt.title("tanh")
plt.show()
```

## 03.Relu(Rectified Linear Unit)
-  y = 0 if (x <= 0) else x

```python
def relu_function(x):
     return np.where(x <= 0, 0, x)

x = np.linspace(-5, 5)
y = relu_function(x)
plt.plot(x,y)
plt.title("relu")
plt.show()
```

## 04. Leaky ReLU
- x가 음수인 영역에서 아주 작은 기울기를 생성 
- 출력이 0이 되어 더 이상 학습이 진행되지 않는 뉴런이 발생하는 dying ReLU 현상을 피하기 위해서 사용한다.
- y = 0.01*x if (x= <=0) else x

```python
def leaky_relu_function(x):
     return np.where(x <= 0, 0.01*x, x)

x = np.linspace(-5, 5)
y = leaky_relu_function(x)
plt.plot(x,y)
plt.title("Leaky relu")
plt.show()
```

## 05. Softmax 
- 입력받은 값을 출력으로 0 ~ 1 사이의 값으로 모두 정규화하여, 출력 값들의 총합은 항상 1이 되는 특성을 가진 함수  
- y = 1 / ( 1 + e<sup>-x</sup>)

```python

def softmax_function(x):
    return np.exp(x) /np.sum(np.exp(x))

y = softmax_function(np.array([1,2,3]))
print(y)
```
