```python
import os 
import pandas as pd 
import tensorflow as tf
from tensorflow.keras import utils 
```

```python
f_path = 'D:/Python_Exercise/Machine_Learning/NLPch03/'
data_set = utils.get_file(fname=f_path + 'imdb.tar.gz', 
                          origin='http://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz',
                          extract = True, cache_subdir = 'datasets',
                         cache_dir = f_path)
```

```python
def directory_data(directory):
    data = {}
    data['review'] = []
    for file_path in os.listdir(directory): #리스트 디렉토리로 넘겨줌 
        with open(os.path.join(directory, file_path), 'r', encoding='utf-8') as file:
            data['review'].append(file.read())
            
    return pd.DataFrame.from_dict(data)
        
```

```python
def data(directory):
    pos_df = directory_data(os.path.join(directory, 'pos'))
    neg_df = directory_data(os.path.join(directory, 'neg'))
    pos_df['sentiment'] = 1
    neg_df['sentiment'] = 0
    
    return pd.concat([pos_df, neg_df])
```

```python
train_df = data(os.path.join(f_path, 'datasets', 'aclImdb', 'train'))
test_df = data(os.path.join(f_path, 'datasets', 'aclImdb', 'test'))
```

```python
print(train_df.tail())
```

```python
reviews = list(train_df['review'])
```

```python
print(reviews[:3])
```

```python
tokenized_reviews = [r.split() for r in reviews]
```

```python
print(tokenized_reviews[0])
```

```python
review_len_by_token = [len(t) for t in tokenized_reviews]
```

```python
print(review_len_by_token[0])
```

```python
print(len(review_len_by_token))
```

```python
review_len_by_eumjeol = [len(s.replace(' ','')) for s in reviews]
```

```python
print(review_len_by_eumjeol[0])
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(10,5))
plt.hist(review_len_by_token, bins = 50, alpha = 0.5, color = 'r', label='length of word')
plt.hist(review_len_by_eumjeol, bins = 50, alpha = 0.5, color = 'b', label='alphabet')
plt.yscale('log', nonpositive='clip')
plt.title('Review Length Histogram')
plt.legend(['word','character'])
plt.xlabel('Review Length')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
import numpy as np 
print('문장 최대 길이 : {}'.format(np.max(review_len_by_token)))
print('문장 최소 길이 : {}'.format(np.min(review_len_by_token)))
print('문장 평균 길이 : {:.2f}'.format(np.mean(review_len_by_token)))
print('문장 길이 표준편차 : {:.2f}'.format(np.std(review_len_by_token)))
print('문장 중간 길이 : {:.2f}'.format(np.median(review_len_by_token)))
print('제1사분위 길이:{}'.format(np.percentile(review_len_by_token, 25)))
print('제3사분위 길이:{}'.format(np.percentile(review_len_by_token, 75)))
```

```python
plt.figure(figsize=(10,5))
plt.boxplot([review_len_by_token], labels=['token'], showmeans=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot([review_len_by_eumjeol], labels=['eumjeol'], showmeans=True)
plt.show()
```
