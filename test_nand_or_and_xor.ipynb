```python
import numpy as np
```

```python
w11 = np.array([-2, -2])
w12 = np.array([2,2])
w2 = np.array([1,1])
b1 = 3 
b2 = -1
b3 = -1
```

```python
# 퍼셉트론
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1
```

```python
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)
```

```python
NAND(0,0)
```

```python
# OR 게이트
def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)
```

```python
OR(0,0)
```

```python
# AND 게이트
def AND(x1, x2):
    return MLP(np.array([x1, x2]), w2, b3)
```

```python
AND(0,0)
```

```python
# XOR 게이트
def XOR(x1, x2):
    return AND(NAND(x1, x2),OR(x1, x2))
```

```python
XOR(1,0)
```

```python
# x1, x2 값을 번갈아 대입하며 최종값 출력
if __name__ == '__main__':
    for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = XOR(x[0], x[1])
        print("입력 값: " + str(x) + " 출력 값: " + str(y))
```
