## Regular Expression 
- 어떤 문자열의 패턴을 쉽게 찾을 수 있도록 하는것이다. 
- 이를 테면 회원가입시 이메일 주소를 입력하기전에 잘못된 주소를 거르기 위해 이메일 패턴을 만들어야 한다. 
    
    정규표현식으로 그러한 과정에 도움을 얻을 수 있다. 

### 대괄호 () or [] : 그 사이의 문자와 매칭하라는 의미

- 대괄호 사이의 문자는 기본적으로 OR조합이다 
- ex. [abc] : 해당 텍스트에 a 또는 b또는 c라는 문자가 있는지 검사하라는 의미가된다. 
- ex. [Tt]om : Tom 또는 tom을 찾으라는 의미가 된다. 

```python
import re
```

```python
p = re.compile(r'[Tt]om')
```

```python
result = p.match('Tom was a trash like man but, uncletom was a pretty good guy')
print(result)
```

```python
result = p.match('My name is Tom')
print(result)
#이경우에는 찾지 못하는데 대문자T또는 소문자t로 시작하는놈을 찾는데 여기서는 앞에 다른놈들이 있기 때문이다. 
```

```python
result = re.match('[A-Za-z][0-9]', 'j12345')
#앞쪽에 패턴을 주고 뒤쪽에 문자열을 주는방식으로 간편하게 쓸수도 있다. 
#첫글자가 A부터 z중에 하나, 두번쨰 글자가 0부터 9까지 중에 하나를 가져와라 
print(result)
```

```python
result = re.match('[A-Za-z][0-9]', 'jk12345')
print(result)
```

### 반복 메타 문자
- 휴대폰 번호  
- 위의 방식으로 휴대전화 번호 매칭 방법  
[0-9][0-9][0-9]-[0-9][0-9][0-9][0-9]-[0-9][0-9][0-9][0-9] : 4자리- 4자리- 4자리

- '+' 한 개 이상 출현  
- [0-9]+-[0-9]+-[0-9]+  
하지만 여전히 문제, 기존 3-3-4 자리 사용하는 번호도 존재

- {} 는 반복 회수를 지정  
- {0,} - 0번 이상 출현  
- {1,} - 1번 이상 출현  
- {1,3} - 1번에서 3번까지 출현


- 위 휴대전화 번호를 아래와 같이 변경  
- [0-9]{3}-[0-9]{3,4}-[0-9]{4}  : 0에서 9까지 3번, 0에서9까지 3~4번, 0에서 9까지 4번 출현

- 국번은 010, 011, 019, 017, 018, 016 까지 존재하는데 .... 이경우는 어떻게 써주냐 하면...
- 01[016789]-[0-9]{3,4}-[0-9]{4}

```python
p = re.compile(r'01[016789]-[0-9]{3,4}-[0-9]{4}')
result = p.match('010-3361-4327')
print(result)
```

### '*' : 0번이상 반복 

```python
p = re.compile(r'abc[0-9]*def')
result = p.match('abc356123123def')
print(result)
```

```python
p = re.compile(r'abc[0-9]*def')
result = p.match('abcdef')
print(result)
```

```python
p = re.compile(r'abc[0-9]*def')
result = p.match('abcddef')
print(result)
```

```python
p = re.compile(r'[^5]abc') #5만 아니면 뭐가와도 상관없다는 뜻이다. 
result = p.match('1abc')
print(result)
```

```python
p = re.compile(r'[^5]abc') #5만 아니면 뭐가와도 상관없다는 뜻이다. 
result = p.match('abc')
print(result)
```

```python
p = re.compile(r'[^5]abc') #5만 아니면 뭐가와도 상관없다는 뜻이다. 
result = p.match('cabc')
print(result)
```

```python
p = re.compile(r'abc[0-9]?def') #나오는것이 0번 또는 1번 등장하는 경우
result = p.match('abcdef')
print(result)
```

```python
p = re.compile(r'abc[0-9]?def') #나오는것이 0번 또는 1번 등장하는 경우
result = p.match('abc2def')
print(result)
```

```python
p = re.compile(r'abc[0-9]?def') #나오는것이 0번 또는 1번 등장하는 경우
result = p.match('abc10def')
print(result)
```

```python
p = re.compile(r'^3[A-Za-z]+9$') 
#대괄호 밖의 ^: 반드시 3으로 시작해야 한다는 의미, 중간에 알파벳이 1개이상나오고 
#맨 마지막은 9로 끝나야 한다는 의미가 된다.                                 
result = p.match('3defr9')
print(result)
```

```python
p = re.compile(r'^3[A-Za-z]+9$') 
result = p.match('34defr9')
print(result)
```

```python
p = re.compile(r'^3[A-Za-z]+9$') 
result = p.match('3defr98')
print(result)
```

```python
p = re.compile(r'^3[A-Za-z]+9$') 
result = p.match('1defr9')
print(result)
```

```python
p = re.compile(r'[0-9].8$') #점은 뭐가오든 1개가 반드시 오고 다음 $은 그 뒷문자로 끝나야 한다는의미 
print(p.match('4p8'))
```

```python
print(p.match('448'))
```

```python
print(p.match('48'))
```

```python
p = re.compile(r'\dabc')
#\d는 숫자 digit을 의미한다. [0-9]와 같은의미지만 갯수가 하나밖에 안된다. 
print(p.match('1abc'))
```

```python
print(p.match('11abc'))
```

```python
print(p.match('9abc'))
```

```python
p = re.compile(r'\D123') #대문자D는 숫자가 아닌놈, 문자가 와야한다는 뜻이된다.
print(p.match('a123'))
```

```python
print(p.match('D123'))
```

```python
print(p.match('1123'))
```

```python
p = re.compile(r'abc\s[0-9]+')
#\s 는 spacebar를 의미한다. 
# +는 갯수를 말한다. 여러개가 와도 좋다는의미
print(p.match('abc 76'))
```

```python
p = re.compile(r'abc\S[0-9]+')
#대문자 \S는 spacebar말고 다른것을 말한다. 
print(p.match('abc 76'))
```

```python
p = re.compile(r'\w[0-9]+')
#\w는 word라는 의미로 알파벳, 또는 숫자가 올수 있다. 
print(p.match('f948484'))
```

```python
p = re.compile(r'\w[0-9]+')
print(p.match('-f948484'))
#알파벳이나 숫자가 아니면 안된다.
```

```python
p = re.compile(r'\W[0-9]+')
#\W , 대문자로 쓰면 알파벳이나 숫자가 오면 안된다. 
print(p.match('-948484'))
```

#### 이메일 정규표현식 
^[\w.%+\-]+@[\w.\-]+\.[A-Za-z]{2,3}$
