```python
##교재의 소스가 과거의 것이므로 그대로 따라하면 에러납니다 !!

import numpy as np 
import matplotlib.pyplot as plt 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import optimizers
```

```python
b_size = 5
train_datagen = ImageDataGenerator(rescale=1./255, 
                                  horizontal_flip=True,
                                  width_shift_range=0.1,
                                  height_shift_range=0.1,
                                  fill_mode = 'nearest')
#rescale : 이미지 자체의 픽셀값을 0에서 1사이로 바꿈
#width shift range : 10퍼센트 범위내에서 좌우로 랜덤하게 이동시킴
#height shift range : 10퍼센트 범위내에서 위아래로 랜덤하게 이동시킴

train_generator = train_datagen.flow_from_directory('train', target_size = (150,150), 
                                                 batch_size = b_size,
                                                 class_mode='binary')
#flow_from_directory로 데이터를 불러올때 첫번째 폴더를 1로 두번째 폴더를 0으로 자동으로 타겟을 잡는다. 
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
#이놈은 학습용이 아니므로 부풀릴 필요가 없다.
test_generator = test_datagen.flow_from_directory('test', target_size = (150,150), 
                                                 batch_size = b_size,
                                                 class_mode='binary')
```

```python
model = Sequential()
model.add(Conv2D(32, (3,3), input_shape=(150, 150, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('softmax'))
#굳이 sigmoid가 아닌 softmax를 쓰는 이유는 저자가 굳이 손실함수로 sparse_categorical_crossenptropy를 
#썻기 떄문이다. 
model.compile(loss='sparse_categorical_crossentropy', 
              optimizer = optimizers.Adam(learning_rate=0.0002),
              #천천히 학습하기 위해 learning_rate을 극도로 낮춘다. 
             metrics = ['accuracy'])
#타겟을 원핫인코딩으로 바꾸지 않은경우에 sparse categorical crossentropy를 써줄수 있다. 
model.summary()
```

```python
steps_train = len(train_generator)
steps_test = len(test_generator)
```

```python
print(steps_train, steps_test)
```

```python
hist = model.fit(train_generator, steps_per_epoch=steps_train, epochs = 20, 
                 validation_data=test_generator, validation_steps=steps_test)

#steps_per_epoch : 배치사이즈를 감안에서 한 에포크당 몇번을 돌아야 하느냐를 써준것이다.
```

```python
acc = hist.history['accuracy']
val_acc = hist.history['val_accuracy']
y_vloss = hist.history['val_loss']
y_loss = hist.history['loss']

x_len = np.arange(len(y_loss))
plt.figure(figsize=(10, 8))
plt.plot(x_len, acc, marker='.', c='red', label='Trainset_acc')
plt.plot(x_len, val_acc, marker='.', c='lightcoral', label='Testset_acc')
plt.plot(x_len, y_vloss, marker='.', c='cornflowerblue', label='Testset_loss')
plt.plot(x_len, y_loss, marker='.', c='blue', label='Trainset_loss')
plt.legend()
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss/acc')
plt.show()
```
