```python
from tensorflow.keras.datasets import mnist 
from tensorflow.keras import utils
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import matplotlib.pyplot as plt
import numpy as np 
import os 
import tensorflow as tf
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32') /255
#만약 컬러가 들어가면 RGB이므로 각 1장씩 3장이 필요한데
#그 때는 3을 써준다. 
#CNN을 쓸떄는 그림을 그대로 집어넣기 때문에 전처럼 찌부시킬필요가 없다. 그냥 차원을 하나 더 늘려준다. 
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32') / 255
Y_train = utils.to_categorical(Y_train)
Y_test = utils.to_categorical(Y_test)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size =(3,3), input_shape=(28,28,1), activation='relu' ))
#Conv2D 2차원 신경망, 그림이 한장 들어오면 출력은 32장의 그림이된다.
#input_dim이 아닌 input_shape이다. 그림한장에 (28,28,1)이다.
model.add(Conv2D(64, kernel_size = (3,3), activation='relu' ))
##이미지 한장을 64개로 만든다는것이다. 32개를 64개로 만드는것이 아니다.... 오해말기...
##그러니까 64 X 32가 되는것이다. 
model.add(MaxPooling2D(pool_size=2))
#maxpooling 특징값은 가져오고 계산량은 줄여준다. 
model.add(Dropout(0.25))
#dropout: 역전파를 계산할 때 은닉층의 일부 노드를 스위치 끄듯이 끈다는 것이다. 
#dropout(0.25)는 25퍼센트를 무작위로 잡아서 꺼버리겠다는것이다.
#다만 이런다고 해서 출력노드의 갯수가 변하는것도 아니다. 
model.add(Flatten())
#flatten dense층에 집어넣기 위해 1차원으로 쫙펴주겠다는것임 
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
# 위 표의 해석은 lecture/cnn_param.xlsx 참조 
```

- 커널이 출력수 만큼있고 그걸 출력수 만큼 곱하면 288 거기에 바이어스 주면 320 이걸 relu함수를 통과시킨다.
- 여기에는 0값도 있고 음수값도 모두 포함되어있는것이다.
- 그래서 마스크로 인해 줄어든 가장자리 값으로 인해 출력값은 계속 줄어든다. 
- 일차원으로 피면 12 X 12 X 64임 
- 여기서 나온 1199882개의 파라미터를 학습시키는것이다. 

```python
model.compile(loss='categorical_crossentropy', 
             optimizer = 'adam', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import os
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
    #해당 폴더의 존재유무를 체크해준다. 

modelpath = './model/{epoch: 02d}-{val_loss:.4f}.hdf5'
checkpointer = ModelCheckpoint(filepath = modelpath, monitor = 'val_loss', verbose=1,
                              save_best_only=True)

early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10)
```

```python
history = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), 
                   epochs=30, batch_size =200, verbose=0, callbacks=[early_stopping_callback,
                                                                     checkpointer])
#validation split할거 없이 이미 split해놓은 test데이터를 넣어주자
```

```python
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
```

```python
y_vloss = history.history['val_loss']
```

```python
y_loss = history.history['loss']
```

```python
x_len = np.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='x', c="red", label='Tested_loss')
plt.plot(x_len, y_loss, marker='.', c="blue", label='Trained_loss')

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
