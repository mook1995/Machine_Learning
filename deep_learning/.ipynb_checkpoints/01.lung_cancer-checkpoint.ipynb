## 예제: 폐암 수술 환자의 생존율 예측하기 실습 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np 
import tensorflow as tf
```

```python
np.random.seed(3)
tf.random.set_seed(3)
#만약 gpu를 사용하는 경우에는 텐서플로가 별도의 내부 랜덤시드를 사용할 수도 있다. 
```

```python
Data_set = np.loadtxt("data/ThoraricSurgery.csv", delimiter=",")
```

```python
X = Data_set[:,:17]
Y = Data_set[:, 17]
```

```python
X.shape
```

```python
print(X[1, :])
```

```python
print(Y)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation ='relu'))
#첫번째 인자는 출력
#Dense 는 Fully Connect Network을 의미한다. 
model.add(Dense(1, activation='sigmoid'))
#여기서는 input_dim을 쓸필요가 없는데 이전층에 이미 써줬기 때문이다.
```

```python
model.summary()
#이걸 본인이 스스로 계산할수 있어야 레이어와 레이어 사이를 어떻게 넘어가는지 이해하고 있다는 것이므로 
#해보길 바람
#hidden layer : w 510개, bias 30개, 
#output layer : w 30개 , bias 1개 
#total : 571개
```

```python
model.compile(loss ='binary_crossentropy', optimizer  ='adam', metrics = ['accuracy'])
model.fit(X, Y, epochs = 100, batch_size =10)
```
