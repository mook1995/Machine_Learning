```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split 
import numpy as np 
import pandas as pd 
import tensorflow as tf
```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(1)
```

```python
df = pd.read_csv('data/housing.csv', delim_whitespace=True)
#데이터 파일을 보면 공백으로 데이터가 구분이 되어있는것을 확인할 수 있는데, 여러칸인것으로 보아 탭으로 보인다.
#이런경우 delim_whitespace=True로 탭이나 모든 공백을 delimiter로 읽어들인다. 
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
dataset = df.values
#이 데이터 역시 float으로 자동으로 잡힐것이므로 데이터 변환없어도 된다, 
X = dataset[:,:13]
Y = dataset[:,13]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=seed)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=13, activation = 'relu'))
model.add(Dense(6, activation ='relu'))
model.add(Dense(1))
#출력에는 활성함수를 안쓸건데 여기에서는 출력값 자체를 사용할것이므로 안써도 된다. 
#분류를 하는 경우에는 sigmoid함수를 사용해서 1또는 0으로 값이 출력되게 한뒤에 퍼센테이지를 계산할수 있지만 
#회귀의 경우에는 바로 모델을 사용한 예측값을 뽑고 그 값을 실제 test데이터의 값과 비교를 하여 
#그래프를 그리거나, 실제와 예측치의 오차를 비교하는등에 사용할 수 있는것이다. 
```

```python
model.summary()
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
model.fit(X_train, y_train, epochs=200, batch_size = 10)
```

- loss : 16.1956은 평균 16천 달러의 오차가 생겼다는 의미가 된다.
- 이전과 같이 분류의 경우 맞힌 클래스의 퍼센테이지로 나오지만 이 경우, 회귀의 경우에는 실제값이 그대로 나온다. 
- 이렇게 예측한 값을 실제test데이터의 값과 비교해보면 정확도를 비교할 수 있다. 
이때 predict를 이용해서 ,... 
모델을 적용한 x_test를 확인하는게 가능다.

```python
y_prediction = model.predict(X_test).flatten() 
#flatten 함수를 이용해 2차원 배열을 1차원으로 납작하게 만들수 있다. 
```

```python
print(y_prediction)
```

```python
for i in range(len(y_prediction)):
    label = y_test[i]
    prediction = y_prediction[i]
    print("실제가격: {:.3f}, 예상가격: {:.3f}".format(label, prediction))
```
